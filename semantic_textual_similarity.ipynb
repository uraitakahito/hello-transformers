```python
text_sim_pipeline = pipeline(model="llm-book/bert-base-japanese-v3-jsts", function_to_apply="none")
text = "川縁でサーフボードを持った人たちがいます"
sim_text = "サーファーたちが川縁に立っています"
# 類似度を計算
result = text_sim_pipeline({"text": text, "text_pair": sim_text})
print(result["score"])
```

```python
dissim_text = "トイレの壁に黒いタオルがかけられています"
result = text_sim_pipeline({"text": text, "text_pair": dissim_text})
print(result["score"])
```
